## Statistics Basics

1 What is statistics, and why is it important?
* Statistics is the branch of mathematics that deals with collecting, organizing, analyzing, interpreting, and presenting data. It helps us make sense of large amounts of information by summarizing it in a meaningful way.

 Statistics Important

5 What is sampling in statistics.
* Sampling in statistics is the process of selecting a small group (sample) from a larger group (population) to study and analyze.

6 What are the different types of sampling methods?
* There are two main types of sampling methods in statistics: probability sampling and non-probability sampling.

1. Probability Sampling
In this method, every individual in the population has a known and equal chance of being selected.

Types:
Simple Random Sampling

Every member is chosen by chance.

Example: Pick 10 names from a box of 100 names.

Systematic Sampling

Select every k-th member.

Example: Every 10th person on a list.

Stratified Sampling

Divide population into groups (strata), then sample from each group.

Example: Divide by gender, then randomly select from each group.

Cluster Sampling

Divide population into clusters, then randomly select whole clusters.

Example: Choose 3 schools randomly, then survey all students in those schools.

2. Non-Probability Sampling
In this method, not everyone has a known or equal chance of being selected.

Types:
Convenience Sampling

Choose people who are easy to reach.

Example: Ask your friends or classmates.

Judgmental/Purposive Sampling

Researcher selects who they think is best for the study.

Example: Choose experts for a special interview.

Snowball Sampling

Existing participants help recruit more participants.

Example: Use one respondent to refer others (common in hidden populations).

Quota Sampling

Select people based on specific quotas (like age, gender).

Example: 50 males and 50 females.

7 What is the difference between random and non-random sampling?
* The difference between random and non-random sampling lies in how samples are chosen from the population:

Random Sampling
Every member of the population has an equal chance of being selected.

It is unbiased and representative.

Used in statistical and scientific studies for accuracy.

Example: Choosing 10 students from a class of 100 using a lottery method.

 Non-Random Sampling
Samples are selected based on judgment, convenience, or other criteria.

It may be biased and not fully representative.

Used in exploratory or practical situations where random sampling isn't possible.

Example: Interviewing only the first 10 people entering a mall.

10  Explain nominal, ordinal, interval, and ratio levels of measurement?
* Here’s a clear explanation of the four levels of measurement in statistics: Nominal, Ordinal, Interval, and Ratio — from the lowest to the highest level:

🔹 1. Nominal Level
Categorical data with no order.

Only used to label variables.

Examples:

Gender: Male, Female

Colors: Red, Blue, Green

Blood type: A, B, AB, O

✅ You can group them but not sort or do math.

🔹 2. Ordinal Level
Categorical data with a meaningful order, but differences aren’t measurable.

Examples:

Education level: High School < Bachelor's < Master's

Satisfaction: Dissatisfied < Neutral < Satisfied

Ranking in a race: 1st, 2nd, 3rd

✅ You can rank them, but can't measure exact difference between them.

🔹 3. Interval Level
Numerical data with equal intervals between values.

No true zero point.

Examples:

Temperature in °C or °F: 10°C, 20°C

IQ scores: 100, 110, 120

Calendar years: 1990, 2000, 2010

✅ You can add and subtract, but not multiply/divide (because zero isn’t absolute).

🔹 4. Ratio Level
Like interval data, but with a true zero point.

Supports all arithmetic operations.

Examples:

Weight: 0 kg, 10 kg

Age: 0 years, 25 years

Income: ₹0, ₹50,000

✅ You can do all math operations: +, –, ×,

```python
2  Write a Python program to compute the variance and standard deviation of a dataset.
*# Program to compute variance and standard deviation

# Sample dataset
data = [10, 12, 23, 23, 16, 23, 21, 16]

# Step 1: Calculate the mean
mean = sum(data) / len(data)

# Step 2: Calculate the squared differences from the mean
squared_diffs = [(x - mean) ** 2 for x in data]

# Step 3: Calculate variance (population)
variance = sum(squared_diffs) / len(data)

# Step 4: Calculate standard deviation
std_deviation = variance ** 0.5
```

9 Write a Python script to calculate covariance between two datasets?
*import numpy as np

# Sample datasets
x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]

# Manual calculation of covariance
def calculate_covariance(x, y):
    if len(x) != len(y):
        raise ValueError("Datasets must be of equal length.")
    
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n

    covariance = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)
    return covariance

# Using NumPy
def numpy_covariance(x, y):
    return np.cov(x, y, ddof=1)[0][1]

# Output results
print("Manual Covariance:", calculate_covariance(x, y))
print("NumPy Covariance:", numpy_covariance(x, y))

```python
12  Implement and compare simple random sampling and systematic sampling?
* import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate a dataset
np.random.seed(42)
population_size = 1000
data = pd.DataFrame({
    'Value': np.random.randint(1, 100, population_size)
})

# Simple Random Sampling
def simple_random_sampling(data, sample_size):
    return data.sample(n=sample_size, random_state=42)

# Systematic Sampling
def systematic_sampling(data, sample_size):
    interval = len(data) // sample_size
    start = np.random.randint(0, interval)  # Random start point
    indices = list(range(start, len(data), interval))[:sample_size]
    return data.iloc[indices]

# Set sample size
sample_size = 100

# Perform sampling
simple_random_sample = simple_random_sampling(data, sample_size)
systematic_sample = systematic_sampling(data, sample_size)

# Compare the samples
print("Simple Random Sample Mean:", simple_random_sample['Value'].mean())
print("Systematic Sample Mean:", systematic_sample['Value'].mean())
print("Population Mean:", data['Value'].mean())

# Visualization
plt.figure(figsize=(10, 5))

# Plot Simple Random Sampling
plt.subplot(1, 2, 1)
plt.hist(simple_random_sample['Value'], bins=10, alpha=0.7, label='Simple Random Sample', color='blue')
plt.axvline(data['Value'].mean(), color='red', linestyle='dashed', label='Population Mean')
plt.title('Simple Random Sampling')
plt.legend()

# Plot Systematic Sampling
plt.subplot(1, 2, 2)
plt.hist(systematic_sample['Value'], bins=10, alpha=0.7, label='Systematic Sample', color='green')
plt.axvline(data['Value'].mean(), color='red', linestyle='dashed', label='Population Mean')
plt.title('Systematic Sampling')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
14 Simulate data using Python and calculate its central tendency and dispersion.
* import numpy as np
import pandas as pd
from scipy import stats

# 1. Simulate data (e.g., 100 random numbers from a normal distribution)
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10

# 2. Convert to pandas Series for convenience
data_series = pd.Series(data)

# Central Tendency
mean = data_series.mean()
median = data_series.median()
mode = data_series.mode().iloc[0]  # take the first mode if multiple

# Dispersion
data_range = data_series.max() - data_series.min()
variance = data_series.var()
std_deviation = data_series.std()

# Display results
print("Central Tendency:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")

print("\nDispersion:")
print(f"Range: {data_range:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")
```

```python
18 Implement Z-score normalization and explain its significance.
* import numpy as np

# Sample data
data = [10, 12, 14, 16, 18]

# Step 1: Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Step 2: Apply Z-score normalization
z_scores = [(x - mean) / std_dev for x in data]

print("Original data:", data)
print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Z-scores:", z_scores)
```

```python
19 Compare two datasets using their standard deviations.
* import numpy as np

# Dataset A: Consistent scores
dataset_A = [50, 52, 49, 51, 50, 50]

# Dataset B: More spread
dataset_B = [40, 60, 30, 70, 50, 50]

# Calculate standard deviations
std_A = np.std(dataset_A)
std_B = np.std(dataset_B)

print("Standard Deviation of Dataset A:", std_A)
print("Standard Deviation of Dataset B:", std_B)

# Interpretation
if std_A > std_B:
    print("Dataset A has more variability.")
elif std_A < std_B:
    print("Dataset B has more variability.")
else:
    print("Both datasets have equal variability.")
```

```python
20 Write a Python program to visualize covariance using a heatmap.
* import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate sample data
np.random.seed(0)
data = {
    'X': np.random.normal(10, 2, 100),
    'Y': np.random.normal(20, 5, 100),
    'Z': np.random.normal(30, 10, 100)
}

df = pd.DataFrame(data)

# Step 2: Calculate the covariance matrix
cov_matrix = df.cov()

# Step 3: Plot heatmap of covariance matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title("Covariance Matrix Heatmap")
plt.show()
```

```python
22 Generate a dataset and implement both variance and standard deviation computations.
*import numpy as np

# Generate a dataset of 10 random integers between 10 and 100
data = np.random.randint(10, 100, size=10)
print("Dataset:", data)

# Calculate Variance
variance = np.var(data)  # Population variance
sample_variance = np.var(data, ddof=1)  # Sample variance

# Calculate Standard Deviation
std_deviation = np.std(data)  # Population standard deviation
sample_std_deviation = np.std(data, ddof=1)  # Sample standard deviation

# Output the results
print("Population Variance:", variance)
print("Sample Variance:", sample_variance)
print("Population Standard Deviation:", std_deviation)
print("Sample Standard Deviation:", sample_std_deviation)
```

```python
23 Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn.
* import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Generate a dataset
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution

# Compute skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data)  # By default, Fisher’s definition (normal == 0)

print(f"Skewness: {data_skewness:.3f}")
print(f"Kurtosis: {data_kurtosis:.3f}")

# Plot histogram and KDE to visualize
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='skyblue', bins=30)
plt.title(f"Histogram with Skewness = {data_skewness:.3f} | Kurtosis = {data_kurtosis:.3f}")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
24 Implement the Pearson and Spearman correlation coefficients for a dataset.
* import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Sample dataset
data = {
    'X': [10, 20, 30, 40, 50],
    'Y': [12, 24, 33, 47, 58]
}

df = pd.DataFrame(data)

# Pearson correlation
pearson_corr, pearson_p = pearsonr(df['X'], df['Y'])

# Spearman correlation
spearman_corr, spearman_p = spearmanr(df['X'], df['Y'])

# Results
print("Pearson Correlation Coefficient:", round(pearson_corr, 4))
print("Pearson p-value:", round(pearson_p, 4))
print("Spearman Correlation Coefficient:", round(spearman_corr, 4))
print("Spearman p-value:", round(spearman_p, 4))
```
